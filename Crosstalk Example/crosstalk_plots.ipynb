```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ell bins from sims
ell_bins = np.load('sim_data/ell_bins.npy')
delta_ell = 50

#camb spectra
ell_lensed = np.loadtxt('camb_spectra/lensed_total_cls_r_0_001.rtf',usecols=0)
TT = np.loadtxt('camb_spectra/lensed_total_cls_r_0_001.rtf',usecols=1)
EE = np.loadtxt('camb_spectra/lensed_total_cls_r_0_001.rtf',usecols=2)
lensedBB = np.loadtxt('camb_spectra/lensed_total_cls_r_0_001.rtf',usecols=3)
ell_BB = np.loadtxt('camb_spectra/total_cls_r_0_001.rtf',usecols=0)
BB = np.loadtxt('camb_spectra/total_cls_r_0_001.rtf',usecols=3)

#binning TT spectrum
binned_TT = []
i=0
while i<len(ell_bins):
    start = np.where(ell_lensed==ell_bins[i]-(delta_ell/2))[0][0]
    end = np.where(ell_lensed==ell_bins[i]+(delta_ell/2))[0][0]
    binned_TT.append(np.average(TT[start:end]))
    i+=1
```
